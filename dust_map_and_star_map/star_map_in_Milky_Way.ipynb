```python
import numpy as np
import healpy as hp
import pandas as pd
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
from astropy import units as u
```

The following graphs are drawn based on the database given in http://galaxymap.org/drupal/node/5. The size of the mark "+" shows the absolute magnitude of stars, larger the size, brighter the star. Since this is based on "real world" data, the magnitudes of stars are obtained after the effect of dust.

```python
star_data = np.transpose(np.genfromtxt('star_data.csv', delimiter = ',', usecols=(2,3,4,8)))
np.shape(star_data)
Longitude = star_data[0][1:5166]
Latitude = star_data[1][1:5166]
Distance = star_data[2][1:5166]
Absolute_Magnitude = star_data[3][1:5166]
```

```python
import pylab as pl

hp.mollview(title="")
hp.graticule()


def star_absmag_bar(x, y, z):#x - longitude, y - latitude, z - absolute magnitude
    n = range(0,5165)
    for i in n:
        hp.projplot(x[i], y[i], marker = '+', markersize = z[i], lonlat=True, coord='G', color = 'red')
    pl.show()
                
    #hp.projplot(x, y, marker = '+', markersize = z, lonlat=True, coord='G')
star_absmag_bar(Longitude, Latitude, np.absolute(Absolute_Magnitude))
```
